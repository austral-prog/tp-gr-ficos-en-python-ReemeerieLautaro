## Duracion de las llamadas de un call center medido en segundos

```python
num_estudiantes = 50


horas_estudio = np.random.uniform(low=1, high=20, size=num_estudiantes)
horas_estudio = np.round(horas_estudio, 1) 

calificaciones = 40 + (horas_estudio * 3) + np.random.normal(loc=0, scale=10, size=num_estudiantes)

calificaciones = np.clip(calificaciones, 0, 100)
calificaciones = np.round(calificaciones, 0) 


plt.figure(figsize=(10, 6)) 

plt.scatter(x=horas_estudio,         
            y=calificaciones,        
            s=50,                    
            c='darkblue',            
            alpha=0.7,               
            edgecolors='black',     
            linewidths=0.8)          

plt.title('Relacin entre Horas de Estudio y Calificacin en el Examen', fontsize=16, fontweight='bold')
plt.xlabel('Horas de Estudio (por semana)', fontsize=12) 
plt.ylabel('Calificacin del Examen (0-100)', fontsize=12) 

plt.grid(True, linestyle='--', alpha=0.6)

plt.xlim(0, 22) 
plt.ylim(-5, 105) 
```
